```python
def output_map(debug=0):
    """
        Category defination
        O = others
        I/B = 2
        LOC/MISC/ORG/PER = 4
        Total number of Combination = 1+2*4 = 9
    """
    if( debug == 1 ):
        print("---------inside output_map function---------\n")
    assert(debug ==0 or debug == 1)
    
    IB=["I","B"]
    ST=["LOC","MISC","ORG","PER"]
    output_type = []
    for i in IB:
        for j in ST:
            output_type.append(i+"-"+j)
    output_type.append("O")
    
    ret = dict()
    idx = 0 
    for i in output_type:
        ret[i] = idx
        if( debug == 1 ):
             print(i+"'s index number is %d"%idx )
        idx = idx+1
    if( debug == 1 ):
        print("\n---------end of inside output_map function---------")
    return ret
```

```python
def create_embedded_matrix(dataAddress=['./../Data/eng.train'],frq = 2,debug=0):
    """Create necessary parsing and for creating of embedded matrix
    Keyword arguments:
    :param dataAddress <list> -- the list of source address of the file for vocabulary (default : './../Data/eng.train')
    :param frq <integer> -- threashold frequencey for choosing a word for embedded matrix (default : 2)
    :param debug <boolean> -- print additional information on dataset (deafult : 0)
    :return: a tuple [ <list,dictionary> ]
            list - <all-possible-word-token-for-embedded-matrix>
            dictionary - <key,value> , <word-to-list, hash integer for each value>\

    dataset format : Word-token POS-tag Phrase-tag NE-tag
                     We extracted only Word-token
    """

    
    if( debug == 1 ):
        print("---------inside create_embeded_matrix function---------\n")

    assert(debug==0 or debug ==1)


    isO   = {}
    all_X = []

    for address in dataAddress:
        print("Accessing dataset : {0}".format(address))
        cnt = 0 
        tot = 0 
        with open(address, 'r') as data:
            lines = data.readlines()
            for line in lines:
                temp = line.split()
                l = len(temp)
                if( l != 4 ):
                    continue
                tot = tot+1 
                if( temp[3] == "O" ):
                    cnt = cnt+1 ;
                if( temp[0] not in isO ):
                    isO[ temp[0] ]  = 0
                if( temp[3] != "O" ):
                    isO[ temp[0] ]  = 1
                all_X.append(temp[0])

        if( debug == 1):
            print("\tTotal data : %d"%tot)
            print("\tTotal Number of **Other** NE =",cnt)
            init_reduce = tot-len(all_X)
            print("\tInitial reduction : %d"%(init_reduce))
            print("\tCurrent total : %d"%len(all_X))


    frqCnt=dict()
    for i in all_X :
        if( i not in frqCnt ):
            frqCnt[i] = 1
        else:
            frqCnt[i] = frqCnt[i]+1

    if( debug == 1):
        print("\nSize of the dictionary(Unique Words) : %d"%len(frqCnt))
        p = set(all_X) ;
        assert( len(p) == len(frqCnt) )


    shrinked_X = []
    for k,v in frqCnt.items():
        if( (v >= frq) and (isO[k] == 1) ):
            shrinked_X.append(k)
    shrinked_X.append("*?*other")

    assert(len(shrinked_X) == len(set(shrinked_X)))

    em_hash = dict()
    idx = 0
    for i in shrinked_X:
        if i not in em_hash:
            em_hash[i] = idx
        idx = idx + 1

    if( debug == 1):
        print("Total item after shrinking by frequency-{0} : {1}".format(frq, len(shrinked_X)))

    if( debug == 1 ):
        print("\n---------end of create_embeded_matrix function---------")

    return (shrinked_X,em_hash)
```
